# Descriptive Statistics of Life Expectancy in Southeast Asia

## Introduction

This report explores the factors influencing life expectancy in 11 Southeast Asian countries: Brunei, Cambodia, Indonesia, Laos, Malaysia, Myanmar, Philippines, Singapore, Thailand, Timor-Leste, and Vietnam. Our analysis covers various indicators such as infant deaths, adult mortality, BMI, income, schooling, GDP, and GDP per capita, to understand their impact on life expectancy in both developed and developing nations within the region.

We start by importing the necessary libraries, including pandas and matplotlib, and reading the data files. Next, we perform data wrangling tasks such as subsetting, renaming, re-indexing, aggregating, merging, and managing data types. Finally, we visualise the cleaned and processed data using matplotlib to identify trends and patterns.

## Importing the libraries

We begin by importing the necessary libraries. We use pandas for reading and wrangling the data, and matplotlib to visualise the data with plots and graphs.

```python
# The 'pandas' library is used to read and wrangle the data in Python
import pandas as pd

# The 'matplotlib' library is used to plot and graph the data in Python
import matplotlib.pylab as plt

# Display graphs inline in the notebook
%matplotlib inline
```

## Reading the files

Next, we’ll load the relevant CSV files into pandas DataFrames.

### GDP Data

For the GDP DataFrame, we only require the following columns: **Unnamed: 3**, which includes the country names, and **Unnamed: 4**, which provides the GDP figures in millions of US dollars for the year 2019.

```python
# Load the GDP data (2019-GDP.csv) from the data directory into a pandas DataFrame
gdp_df = pd.read_csv('data/2019-GDP.csv')
gdp_df
```

### Population Data

For the Population DataFrame, we only require the following columns: **Unnamed: 2**, which includes the country names, and **Unnamed: 76**, which provides the population figures in thousands for the year 2019.

```python
# Load the population data (2020-Population.csv) from the data directory into a pandas DataFrame
population_df = pd.read_csv('data/2020-Population.csv')
population_df
```

### Life Expectancy Data

In the Life Expectancy DataFrame, we retain the columns that pertain to country, status, life expectancy, infant deaths, adult mortality, BMI, income composition of resources, and schooling.

```python
# Load the life expectancy data (LifeExpectancyData-v2.csv) from the data directory into a pandas DataFrame
life_expectancy_df = pd.read_csv('data/LifeExpectancyData-v2.csv')
life_expectancy_df
```

## Wrangling the data

Since our analysis focuses on Southeast Asian countries, we need to subset the data from each DataFrame to only include these countries. To do this, we create a list of Southeast Asian countries, taking into account any variations in their representation across different files.

```python
# A list containing the names of all Southeast Asian countries with different representations between the three files
sea_countries = ['Brunei Darussalam', 'Cambodia', 'Indonesia', "Lao People's Democratic Republic", 'Lao PDR', 'Malaysia', 
                 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Timor-Leste', 'Vietnam', 'Viet Nam']
```

### GDP Data

Next, we subset the columns containing the Southeast Asian countries and GDP data for 2019 from the GDP DataFrame. Additionally, we rename these columns for clarity and consistency.

```python
# Extract the data for Southeast Asian countries
gdp_sea = gdp_df[gdp_df['Unnamed: 3'].isin(sea_countries)]

# Reset index
gdp_sea = gdp_sea.reset_index()

# Select the country ('Unnamed: 3') and GDP ('Unnamed: 4') columns
gdp_sea = gdp_sea[['Unnamed: 3', 'Unnamed: 4']]

# Rename the columns
gdp_sea.rename(columns={'Unnamed: 3':'Country', 'Unnamed: 4':'GDP (2019)'}, inplace=True)

# Rename certain countries in order to be suitable to merge with other data frames
gdp_sea['Country'] = gdp_sea['Country'].replace({'Brunei Darussalam':'Brunei', "Lao PDR": 'Laos'})

# Print gdp_sea
gdp_sea
```

### Population Data

Similarly, we subset the columns containing the Southeast Asian countries and population data for 2019 from the Population DataFrame. Additionally, we rename these columns for clarity and consistency.

```python
# Extract the data for Southeast Asian countries
population_sea = population_df[population_df['Unnamed: 2'].isin(sea_countries)]

# Reset index
population_sea = population_sea.reset_index()

# Select the country ('Unnamed: 2') and population of 2019 ('Unnamed: 76') columns
population_sea = population_sea[['Unnamed: 2', 'Unnamed: 76']]

# Rename the columns
population_sea.rename(columns={'Unnamed: 2':'Country', 'Unnamed: 76':'Population (2019)'}, inplace=True)

# Rename certain countries in order to be suitable to merge with other data frames
population_sea['Country'] = population_sea['Country'].replace({'Brunei Darussalam':'Brunei', 'Viet Nam':'Vietnam', 
                                                               "Lao People's Democratic Republic":'Laos'})

# Print population_sea
population_sea
```

### Life Expectancy Data

Similarly, we subset the columns containing the Southeast Asian countries and aggregate the life expectancy, alcohol consumption, BMI, and schooling columns to calculate their mean and maximum values over the period from 2000 to 2015 from the Life Expectancy DataFrame. Additionally, we rename these columns for clarity and consistency.

```python
# Extract the data for Southeast Asian countries
life_expectancy_sea = life_expectancy_df[life_expectancy_df['country'].isin(sea_countries)]

# Aggregate life expectancy, BMI, income composition of resources, and schooling
life_expectancy_sea = life_expectancy_sea.groupby(['country', 'Status']).agg({'Life expectancy ':['mean','max'], 
                                                                              'Alcohol consumption':'mean',
                                                                              ' BMI ':'mean', 
                                                                              'Schooling':'mean'})

# Turn 'country' into column values
life_expectancy_sea = life_expectancy_sea.reset_index()

# Drop the top level in the column hierarchy
life_expectancy_sea.columns = life_expectancy_sea.columns.droplevel(0)

# Rename the columns
life_expectancy_sea.columns = ['Country', 'Status', 'Mean Life Expectancy', 'Max Life Expectancy', 'Mean Alcohol Consumption', 
                               'Mean BMI', 'Mean Schooling']

# Rename certain countries in order to be suitable to merge with other data frames
life_expectancy_sea['Country'] = life_expectancy_sea['Country'].replace({'Brunei Darussalam':'Brunei', 'Viet Nam':'Vietnam',
                                                                         "Lao People's Democratic Republic":'Laos'})

# Print life_expectancy_sea
life_expectancy_sea
```

### Merging DataFrames

We then merge the cleaned DataFrames (GDP, Population, and Life Expectancy) based on the country names.

```python
# Merge life_expectancy_df, population_df and gdp_df
merged_df = pd.merge(life_expectancy_sea, population_sea, on=['Country'])
merged_df = pd.merge(merged_df, gdp_sea, on=['Country'])

# Print merged_df
merged_df
```

### Managing Data Types

In the current DataFrame, the GDP (2019) and Population (2019) columns are stored as objects rather than numeric types, as shown below.

```python
merged_df.dtypes
```

Additionally, the population values are recorded in thousands and formatted using spaces, while GDP values are in millions and formatted using commas. To ensure accurate feature engineering for the GDP per Capita column, we need to:

1. Remove spaces and commas from the respective columns.
2. Convert these columns to numeric data types.
3. Adjust the scale of the values by multiplying the Population values by 1,000 and the GDP values by 1,000,000.
4. Calculate the GDP per Capita column based on these adjusted values.

```python
# Remove any spaces from the numbers in 'Population' column
merged_df['Population (2019)'] = merged_df['Population (2019)'].replace(' ','', regex=True)

# Cast from object to integer data type
merged_df['Population (2019)'] = pd.to_numeric(merged_df['Population (2019)'])

# Multiply all values by 1000 as the population is in thousands
merged_df['Population (2019)'] = merged_df['Population (2019)'].apply(lambda x: x*1000)

# Remove any commas from the numbers in 'GDP' column
merged_df['GDP (2019)'] = merged_df['GDP (2019)'].replace(',','', regex=True)

# Cast from object to integer data type
merged_df['GDP (2019)'] = pd.to_numeric(merged_df['GDP (2019)'])

# Multiply all values by 1000000 as the GDP is in in millions
merged_df['GDP (2019)'] = merged_df['GDP (2019)'].apply(lambda x: x*1000000)

# Add new column 'GDP per Capita'
merged_df['GDP per Capita (2019)'] = merged_df['GDP (2019)'] / merged_df['Population (2019)']

# Print merged_df
merged_df
```

## How does life expectancy differ between developed and developing Southeast Asian countries?

This analysis examines the differences in life expectancy between developed and developing Southeast Asian countries by using a grouped bar chart to compare the average life expectancy for each country from 2000 to 2015.

```python
# Plot grouped bar chart
plt.figure(figsize=(12, 6))

# Group by 'status' to plot each group separately
for status in merged_df['Status'].unique():
    subset = merged_df[merged_df['Status'] == status]
    plt.bar(subset['Country'], subset['Mean Life Expectancy'], label=status)

# Set 'Mean Life Expectancy by Country (2000-2015) for Southeast Asian Countries' as the title
plt.title('Mean Life Expectancy by Country (2000-2015) for Southeast Asian Countries')

# Set 'Country' as the label for the x axis
plt.xlabel('Country')

# Set 'Mean Life Expectancy' as the label for the y axis
plt.ylabel('Mean Life Expectancy')

# Add a legend for the Status of the countries
plt.legend(title='Status')

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Set the y-axis limits to a range that encompasses the mean life expectancy data
plt.ylim(60, 85) 

# Display the grouped bar chart
plt.show()
```

The grouped bar chart shows that Singapore, the only developed country in Southeast Asia, has the highest mean life expectancy compared to the developing countries in the region.

## What effect does infant and adult mortality rates have on life expectancy?

This analysis examines trends in life expectancy, adult mortality, and infant deaths in Singapore through two separate line graphs. The first graph visualises life expectancy over time, while the other graph visualises adult mortality and infant deaths over time.

```python
# Extract the data for Singapore
singapore_df = life_expectancy_df[life_expectancy_df['country']=='Singapore']

# Turn 'Status' into column values
singapore_df = singapore_df.reset_index()

# Select the country, year, infant deaths and adult mortality columns
singapore_df = singapore_df[['country', 'Year', 'Life expectancy ', 'infant deaths', 'Adult Mortality']]

# Rename the columns
singapore_df.rename(columns={'country':'Country', 'Life expectancy ':'Life Expectancy', 
                             'infant deaths':'Infant Deaths'}, inplace=True)

# Print singapore_df
singapore_df
```

```python
# Set the x axis as the year column and the y axis as the life expectancy column
plt.plot(singapore_df['Year'], singapore_df['Life Expectancy'])

# Set 'Life Expectancy in Singapore over Time' as the title of the line graph
plt.title('Life Expectancy in Singapore over Time')

# Set 'Time (Years)' as the label for the x axis
plt.xlabel('Time (Years)')

# Set 'Life Expectancy (Age)' as the label for the y axis
plt.ylabel('Life Expectancy (Age)')

# Display the line graph
plt.show()
```

```python
# Set the x axis as the year column and the y axis as the number of adult mortality column
plt.plot(singapore_df['Year'], singapore_df['Adult Mortality'])

# Set the x axis as the year and the y axis as the number of infant deaths
plt.plot(singapore_df['Year'], singapore_df['Infant Deaths'])

# Set 'Adult Mortality and Infant Deaths in Singapore over Time' as the title of the line graph
plt.title('Adult Mortality and Infant Deaths in Singapore over Time')

# Set 'Time (Years)' as the label for the x axis
plt.xlabel('Time (Years)')

# Set 'Number of Deaths' as the label for the y axis
plt.ylabel('Number of Deaths')

# Display the line graph
plt.show()
```

The life expectancy line graph shows a steady increase in the life expectancy of Singaporeans over time, which might indicate improvements in public health and living conditions, with a notable spike in 2006 before returning to a steady upward trend. 

On the other hand, the adult mortality line graph shows a steady decline in adult mortality rates in Singapore, which might also indicate improvements in public health and living conditions, with a notable dip in 2011, before resuming a steady downward trend. On the other hand, the infant mortality rate remains consistently at zero throughout the years, indicating a successful elimination of infant deaths, likely due to high standards of maternal and child healthcare.

The analysis demonstrates a clear relationship between mortality rates and life expectancy: as infant deaths and adult mortality decrease, life expectancy increases. This correlation is consistent with the understanding that lower mortality rates are indicative of better healthcare, improved living conditions, and effective public health policies, all of which contribute to longer life expectancy. These findings might suggest that continued efforts to reduce adult mortality and maintain low infant mortality rates are crucial for sustaining and improving life expectancy in Singapore.
